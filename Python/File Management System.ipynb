# <u> WEB SCRAPPING

```python
# Imported necessary libraries for both Web Scrapping and file management system
import os
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import shutil
```

```python
# We are going to extract extensions from below site.
url = 'http://www.vidyagyaan.com/computer-knowledge/list-of-computer-file-extensions-and-their-meaning/'
```

```python
# Checking if we have the permission to scrapp this site.
page = requests.get(url)
page
```

```python
# Getting the content inside the page.
soup = BeautifulSoup(page.text)
```

```python
soup
```

```python
# Searching for the required content i.e. extensions using find_all
# Using find_all to gather all the tables
table = soup.find_all('table')
table
```

```python
# Empty list to stored the extracted data
extensions = []
l = 1

for i in table:
    
    # Again, Using find_all to get the table data present in each table.
    cols = i.find_all('td')
    # print(l,len(cols))
    
    # Get text from all the table data
    raw = [e.text for e in cols]
    # print(raw)
    
    # Extensions are present in the even position of each table
    index = [n for n in range(0,len(cols),2) if n>0]
    # print(l,index)
    
    # Getting extensions from each tables.
    ext = list(map(lambda x: raw[x], index))
    # print(l,ext)
    l+=1
    
    # Appending extensions to the empty list
    extensions.append(ext)
extensions
```

```python
# Creating a function to separate nested list
def list_sep(ext):
    new_ext = []
    for element in ext:
        if type(element) is list:
            for item in element:
                new_ext.append(item)
        else:
            new_ext.append(element)
    return new_ext

# Calling list_sep() and assigning a variable ext.
ext = list_sep(extensions)
ext
```

```python
# Cleaning the data.(Removing or, and)

def Clean(x):
    new_list = []
    for i in x:
        x = re.findall('or', i)
        y = re.findall('and',i)
        if x:
            i = i.split(' or ')
            new_list.append(i)
        elif y:
            i = i.split(' and ')
            new_list.append(i)
        else:
            new_list.append(i)
    return new_list

# Making two separate list f
new_ext = Clean(ext)
```

```python
print(len(new_ext))
new_ext
```

```python
# Again, separating nested list
new_ext = list_sep(new_ext)
print(len(new_ext))
new_ext
```

```python
# Removing '.' from the extensions
ext0 = []
for i in new_ext:
    i = i.split('.')
    ext0.append(i[1])
```

```python
# Adding few more extensions
ext0 = ext0 + ['zip'] + ['iso'] + ['ipynb'] + ['ipynb_checkpoints'] + ['webp']
ext0
```

```python
# Saving the file in csv format
ext2 = ext0
extension = pd.DataFrame(ext2)
extension.to_csv('Extensions.csv',index=False)
```

```python
def files(path):
    # change the location of directory
    os.chdir(path)
    # files in the directory
    return os.listdir()

# Enter the files location

path = input('Enter the location of the files: ')

all_files = files(path)
```

```python
def get_files(files):
    list_extension = []

    for file in files:
        # Splitting using . and getting files extensions
        extension = file.split('.')
        if extension[-1] in ext:
            list_extension.append(extension[-1])
            
    return list_extension

# sort files and folders.

sorted_files = get_files(all_files)
```

```python
print(sorted_files)
```

```python
# find the unique extensions present in the list
sorted_files = set(sorted_files)
print(sorted_files)
```

```python
def transfer_files(path, files):
    try:
        shutil.rmtree(path)
        os.mkdir(path)
    except:
        os.mkdir(path)

    try:
        for ex in files:
            print(ex, end=' | ')
            os.mkdir(path + '\\' + ex)
            for file in os.listdir():
                if ex in file:
                    shutil.copy(file, path + '\\' + ex)
    except PermissionError:
        print("Folder cannot transfer.",end=" | ")
        
# Enter where do you want to transfer the files.

path = input("Transfer this files to: " )

transfer_files(path, sorted_files)
```
