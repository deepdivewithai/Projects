```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cars.csv',na_values='?')
df
```

```python
df = df.drop('symboling',axis=1)
```

```python
df.isnull().sum()
```

```python
for i in df.columns:
    if df[i].isnull().sum():
        df[i] = df[i].fillna(df[i].mean())
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ob_col = df.select_dtypes('object').columns
ob_col
```

```python
oe = OrdinalEncoder()
```

```python
df.loc[:,ob_col] = oe.fit_transform(df.loc[:,ob_col])
```

```python
df.head()
```

```python
df.corr().round(4)
```

```python
df = df.drop(['width','height'],axis=1)
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
```

```python
reg = LinearRegression()
```

```python
reg.fit(xtrain,ytrain)
```

```python
reg.score(xtrain,ytrain)
```

```python
reg.score(xtest,ytest)
```

# Hyperparameter Testing

```python
# using random state
```

```python
for i in range(500):
    xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=i)
    reg.fit(xtrain,ytrain)
    print('for random state:',i,'\n')
    if reg.score(xtrain,ytrain) > 0.81 and reg.score(xtest,ytest) > 0.81:
        print('Train:',reg.score(xtrain,ytrain))
        print('Test: ',reg.score(xtest,ytest))
        print()
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=4177)
reg.fit(xtrain,ytrain)
print(reg.score(xtest,ytest))
print(reg.score(xtrain,ytrain))
```

```python
rig = Ridge()
rig.fit(xtrain,ytrain)
print('Train:',rig.score(xtrain,ytrain))
print('Test: ',rig.score(xtest,ytest))
```

```python
lasso = Lasso()
lasso.fit(xtrain,ytrain)
print('Train:',lasso.score(xtrain,ytrain))
print('Test: ',lasso.score(xtest,ytest))
```

```python
for i in range(1,30):
    xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=355)
    rig = Ridge(alpha=i)
    rig.fit(xtrain,ytrain)
    print('for alpha:',i,'\n')
    print('Train:',rig.score(xtrain,ytrain))
    print('Test: ',rig.score(xtest,ytest))
    print()
```

```python
rig = Ridge(alpha=0.1)
rig.fit(xtrain,ytrain)
print('for alpha:',0.1,'\n')
print('Train:',rig.score(xtrain,ytrain))
print('Test: ',rig.score(xtest,ytest))
```

```python
for i in range(1,100):
    xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=234)
    lasso = Lasso(alpha=i)
    lasso.fit(xtrain,ytrain)
    print('for alpha:',i,'\n')
    print('Train:',lasso.score(xtrain,ytrain))
    print('Test: ',lasso.score(xtest,ytest))
    print()
```

```python
lasso = Lasso(alpha=19.5)
lasso.fit(xtrain,ytrain)
print('for alpha:',19.5,'\n')
print('Train:',lasso.score(xtrain,ytrain))
print('Test: ',lasso.score(xtest,ytest))
```
