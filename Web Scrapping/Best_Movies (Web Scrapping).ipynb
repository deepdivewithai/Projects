```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings('ignore')
```

```python
url = 'https://www.timeout.com/film/best-movies-of-all-time'
```

```python
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.text)
soup
```

```python
content = soup.find('div', class_='_zoneItems_882m9_1 zoneItems')
# content.find_all('article')[0].find('h3').text.split(':')[1].strip()
content = content.find_all('article')
content
```

```python
movies = [i.find('h3').text.replace('\xa0','').split('.')[1] for i in content]
movies
```

```python
first_movie = [i.split(':')[1].strip() for i in movies if i == movies[0]]
movies[0] = first_movie[0]
movies
```

```python
import re
mylist = movies # init the list
year = []
for y in mylist:
    match = re.match(r'.*([1-3][0-9]{3})', y)
    if match is not None:
        year.append(match.group(1))
    else:
        year.append('no year')
```

```python
year
```

```python
movies_list = []
for i in range(len(movies)):
    if i in [50, 52, 95]:
        movies_list.append(movies[i])
        print(i,movies[i])
        continue
    else:
        name = movies[i].replace(str([movies[i].split()[-1]][0]),'').strip()
        movies_list.append(name)
        print(i,name)
```

```python
data = {
    'Rank': range(1,101),
    'Movies': movies_list,
    'Release_Year': year
}
```

```python
df = pd.DataFrame(data)
```
